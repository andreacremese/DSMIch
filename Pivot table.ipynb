# pivot tables

Summarizing data in a data frame. Heavy use of aggregation functoin.
import pandas as pd
import numpy as nptends to include marginal values, or sums 

```python
import pandas as pd
import numpy as np
df = pd.read_csv('cars.csv')
df.head()
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=np.mean)
```
